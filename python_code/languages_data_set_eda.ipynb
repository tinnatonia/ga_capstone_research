```python
def aes_assignment(x):
        if x == 1:
            return 'Not Endangered'
        if x == 2: 
            return 'Vulnerable'
        if x == 3:
            return 'Shifting'
        if x == 4:
            return 'Moribund'
        if x == 5:
            return 'Nearly Extinct'
        if x == 6:
            return 'Extinct'
        else:
            return 'no status'

languages['aes'] = languages['aes'].apply(aes_assignment)
```
